```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('cat_single.jpg',target_size=(400,400))
```

```python
img
```

```python
generate_data = ImageDataGenerator(
    rotation_range = 40,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip = True,
    width_shift_range=0.2,
    height_shift_range=0.2
)
```

```python
type(img)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
in_batch = img.reshape(1,400,400,3)
```

```python
i=0
# flow is for one image to all 
# flow_to_directory to generate a collection of different images
for out in generate_data.flow(in_batch,batch_size=1,save_to_dir='new_cat_generated_data'):
    i+=1
    if(i==15):
        break
```
