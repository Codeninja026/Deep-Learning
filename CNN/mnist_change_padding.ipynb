## with padding='same'

```python
model1 = Sequential()
```

```python
model1.add(Conv2D(30,kernel_size=(3,3), padding='same',activation='relu',input_shape=(28,28,1)))
model1.add(Conv2D(30,kernel_size=(3,3), padding='same',activation='relu',))

model1.add(Conv2D(30,kernel_size=(3,3), padding='same',activation='relu'))

model1.add(Flatten())
model1.add(Dense(128,activation="relu"))
model1.add(Dense(10,activation="softmax"))

```

```python
model1.compile(loss='sparse_categorical_crossentropy',optimizer="adam",metrics=['accuracy'])
```

```python
hist_model1 = model1.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=5)
```

```python
plt.plot(hist.history['val_accuracy'],label="model_val_accuracy")
plt.plot(hist_model1.history['val_accuracy'],label="model1-val_accuracy")
plt.legend()
plt.show()
```
