```python
import numpy as np
import pandas as pd
```

```python
import random
```

```python
class My_percep:
    def __init__(self,epochs=1000,learning_rate=0.01):
        self.epochs = epochs
        self.learning_rate = learning_rate

    def step(self,x):
        if x<=0:
            return 0
        else:
            return 1
    def coef_return(self,X,y):
        X = np.insert(X,0,1,axis=1)
        weights = np.ones(X.shape[1])
        for i in range(self.epochs):
            random_point = random.randint(0,X.shape[0]-1)
            y_hat = self.step(np.dot(weights,X[random_point]))
            weights = weights + self.learning_rate * (y[random_point]-y_hat)*X[random_point]

        return weights[0],weights[1:]
```

```python
clf = My_percep()
intercept,coef = clf.coef_return(X,y)
```

```python
print(intercept)
print(coef)
```

```python
slope = - coef[0]/coef[1]
c = - intercept / coef[1]
```
