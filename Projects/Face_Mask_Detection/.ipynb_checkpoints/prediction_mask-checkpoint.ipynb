```python
import tensorflow
from tensorflow import keras
from keras.models import load_model
import cv2
```

```python
model1 = load_model('model.h5')
```

```python
import matplotlib.pyplot as plt
```

```python
img= cv2.imread('train/with_mask/with_mask_1.jpg')
img = cv2.resize(img,(244,244))
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
from tensorflow.keras.utils import img_to_array
```

```python
import os
print(os.getcwd())  
```

```python
img_array = img_to_array(img)
```

```python
img_array = img_array / 255.0
```

```python
img_array = img_array.reshape(1,244,244,3)
```

```python
model1.predict(img_array)
```

```python
img2= cv2.imread('train/without_mask/without_mask_21.jpg')
img2= cv2.resize(img2,(244,244))
```

```python
img_array2 = img_to_array(img2)
img_array2 = img_array2/ 255.0
img_array2 = img_array2.reshape(1,244,244,3)
model1.predict(img_array2)
```
