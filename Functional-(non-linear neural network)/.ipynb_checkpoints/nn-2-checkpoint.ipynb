```python
import tensorflow
import keras
```

```python
from keras.layers import *
```

```python
from keras import Model
```

```python
inp1 = Input(shape=(3,))
inp2 = Input(shape=(3,))
```

```python
y1 = Dense(256,activation="relu",name="left-hidden-1")(inp1)
y2 = Dense(128,activation='relu',name="left-hidden-2")(y1)
y3 = Dense(64,activation="relu",name="left-hidden-3")(y2)
```

```python
x1 = Dense(128,activation='relu',name="right-hidden-1")(inp2)
x2 = Dense(64,activation='relu',name="right-hidden-2")(x1)
```

```python
combined = concatenate([x2,y3])
```

```python
z = Dense(32,activation='relu',name="combined-hidden-1")(combined)
```

```python
z1 = Dense(1,activation='linear')(z)
```

```python
model = Model(inputs=[inp1,inp2],outputs=z1)
```

```python
from keras.utils import plot_model
```

```python
plot_model(model, show_shapes=True)
```
