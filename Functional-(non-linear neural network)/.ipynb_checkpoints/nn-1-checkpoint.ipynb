```python
import numpy as np
import pandas as pd
```

```python
import tensorflow
from tensorflow.keras.layers import *
```

```python
# suppose if we have 3 columns as input and 2 outputs 
```

```python
inp = Input(shape=(3,))
```

```python
hidden1 = Dense(128,activation='relu',name='hidden1')(inp)
```

```python
hidden2 = Dense(64,activation='relu',name='hidden2')(hidden1)
```

```python
out1 = Dense(1,activation='linear',name="out1")(hidden2)
out2 = Dense(1,activation='sigmoid',name='out2')(hidden2)
```

```python
from keras import Model
```

```python
model = Model(inputs=inp,outputs=[out1,out2])
```

```python
from keras.utils import plot_model
```

```python
plot_model(model, show_shapes=True,dpi=50)
```
