```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)
```

```python
print(X.shape)
print(y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap="RdYlGn")
```

```python
def percep(X,y):

    w1=1
    w2=1
    b=1

    lr =0.1
    for i in range(1000):
        for j in range(X.shape[0]):

            re = w1* X[j][0]  + w2 * X[j][1] + b

            if re*y[j]<0:
                w1 = w1 + lr * y[j]*X[j][0]
                w2 = w2 + lr * y[j]*X[j][1]
                b = b + lr * y[j]

    return w1,w2,b
```

```python
w1,w2,b = percep(X,y)
```

```python
print(f"w1: {w1}")
print(f"w2: {w2}")
print(f"b: {b}")
```

```python
slope =  -(w1/w2)
c = - (b/w2)
print(slope)
print(c)
```

```python
x_input = np.linspace(-4,4,100)
y_input = slope * x_input + c
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap="RdYlGn")
plt.plot(x_input,y_input,c="black")
plt.ylim(-3,2)
```
