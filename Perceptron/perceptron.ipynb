```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100,n_features=2,n_informative=2,n_redundant=0,class_sep=1.0, 
                         n_clusters_per_class=1,n_classes=2,random_state=30)
```

```python
df = pd.DataFrame(X,columns=['col1','col2'])
df['target'] = y
df
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=y,cmap='RdYlGn')
```

```python
# cgpa	iq	placement
sns.scatterplot(x=df.iloc[:,0],y=df.iloc[:,1],hue=df['target'])
```

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X=X,y=y,clf=p)
```
