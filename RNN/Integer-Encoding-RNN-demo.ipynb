```python
import numpy as np
import pandas as pd
```

```python
sentences = [
    "I love my India",
    "She loves my voice",
    "I love her voice",
    "My India is a secure country",
    "Her voice is beautiful"
]
```

```python
from tensorflow import keras
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# oov = out of vocabulary
token = Tokenizer(oov_token="nothing")
```

```python
token.fit_on_texts(sentences)
```

```python
token.word_index
```

```python
token.word_counts
```

```python
seq = token.texts_to_sequences(sentences)
seq
```

```python
from keras.utils import pad_sequences
```

```python
seq = pad_sequences(seq,padding='post')
```
